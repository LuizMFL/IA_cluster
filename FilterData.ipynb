```python
import pandas as pd
import numpy as np
```

```python
movie = pd.read_csv("datasets/rotten_tomatoes_movies.csv")
critic = pd.read_csv("datasets/rotten_tomatoes_critic_reviews.csv")
```

```python
movie_novo = movie.drop(columns=["movie_info", "critics_consensus", "content_rating", "original_release_date", "streaming_release_date", "runtime", "production_company", "tomatometer_status", "tomatometer_rating", "audience_status", "audience_rating", "audience_count", "tomatometer_top_critics_count", "tomatometer_fresh_critics_count", "tomatometer_rotten_critics_count"])
movie_novo
```

```python
critic_novo = critic.drop(columns=["top_critic", "publisher_name", "review_type", "review_date", "review_content"])
critic_novo
```

## Tratamento de dataset 1

```python
#Padronizou o dataset com votos maiores que 100, totalizando 3451
mask_movie = (movie_novo["tomatometer_count"] > 200)
movie_mask = movie_novo.loc[mask_movie]
movie_mask['movie_title']
```

```python
movie_mask.reset_index(drop=True,inplace=True)
movie_mask
```

## Fim do tratamento dataset1

```python
critic_novo_sem_nan = critic_novo.dropna(subset=['critic_name'])
critic_novo_sem_nan
```

```python
dataframe_novo = pd.DataFrame(columns=['rotten_tomatoes_link','critic_name','review_score'])
```

```python
lista = []
for i in movie_mask.rotten_tomatoes_link.values:
    mask_critic = (critic_novo_sem_nan["rotten_tomatoes_link"] == i)
    critic_mask = critic_novo_sem_nan.loc[mask_critic]
    lista.append(critic_mask)
dataframe_novo = pd.concat(lista, ignore_index=True)
dataframe_novo
```

```python
dataframe_novo_atual = dataframe_novo[~dataframe_novo.critic_name.isin(dataframe_novo.critic_name.value_counts().loc[lambda x: (x < 100) ].index)]
dataframe_novo_atual.reset_index(drop=True,inplace=True)
dataframe_novo_atual
```

```python
dataframe_novo_atual.groupby(['critic_name']).count()
```

```python
dataframe_novo_atual.groupby(['review_score']).count()
```

```python
import math
for index, i in dataframe_novo_atual.iterrows():
    if isinstance(i["review_score"], float):
        if not math.isnan(i["review_score"]):
            valor = int(i["review_score"])
            while valor > 10:
                valor = valor // 10
            i["review_score"] = valor
    elif isinstance(i["review_score"], int):
        valor = i["review_score"]
        while valor > 10:
            valor = valor // 10
        i["review_score"] = valor
    elif i["review_score"] == "A+":
        i["review_score"] = 10
    elif i["review_score"] == "A":
        i["review_score"] = 9
    elif i["review_score"] == "A-":
        i["review_score"] = 8
    elif i["review_score"] == "B+":
        i["review_score"] = 7
    elif i["review_score"] == "B":
        i["review_score"] = 6
    elif i["review_score"] == "B-":
        i["review_score"] = 5
    elif i["review_score"] == "C+":
        i["review_score"] = 4
    elif i["review_score"] == "C":
        i["review_score"] = 3
    elif i["review_score"].count("C") and i["review_score"].count("-"):
        i["review_score"] = 2
    elif i["review_score"].count("D"):
        i["review_score"] = 1
    elif i["review_score"].count("F"):
        i["review_score"] = 0
    elif i["review_score"].__contains__("/"):
        valor = i["review_score"].split("/")
        valor1 = int(float(valor[0]))
        valor2 = int(float(valor[1]))
        while valor1 > 10:
            valor1 = valor1 // 10
        while valor2 > 10:
            valor2 = valor2 // 10
        final = (valor1*10)/valor2
        while final > 10:
            final = final // 10
        i["review_score"] = int(final)
    elif isinstance(i["review_score"], str) and i["review_score"].isdigit():
        valor = int(i["review_score"])
        while valor > 10:
            valor = valor // 10
        i["review_score"] = int(valor)
```
